-> The "Labeled Boxes"

"Imagine you are moving houses. You don't just throw everything into a pile; you put things in boxes and write a label on the outside like 'Kitchen' or 'Books.' In Python, a Variable is just a labeled box that holds information so we can find it later."

Rules for Naming Boxes:

Must start with a letter or underscore (e.g., student_name is good, 1st_student is bad).

Case-sensitive (Age and age are different boxes!).

```python
# Creating a "box" named 'gift' and putting a "PlayStation" inside
gift = "PlayStation"

# Changing what's inside the box (re-assignment)
gift = "Bicycle"

print(gift)
```

Floats: Numbers with decimals (like a price: 99.99).

-> "The X-Ray Machine"

# Module 3: Strings & Operators

## 1) Strings:




"The Bead Necklace"

"Think of a String like a necklace made of alphabet beads. Each bead has a specific position. We can look at one bead at a time (Indexing) or cut a section of the necklace to make a bracelet (Slicing)."



```python
#Creation and Indexing
# A string is just text in quotes
necklace = "PYTHON"

# Indexing: Finding a specific "bead" (Computer starts counting at 0!)
print(necklace[0])  # Output: P
print(necklace[-1]) # Output: N (The last bead)
```

```python
# Slicing: [Start : Stop before]
# Let's grab "PYTH"
print(necklace[0:4])

# Let's grab from index 2 to the end ("THON")
print(necklace[2:])
```

## 1.1) String Methods: "The Magic Wand"

Methods are like magic spells we cast on our strings to change their appearance without rewriting them.

```python
shout = "hello world"
print(shout.upper())  # HELLO WORLD


clean_me = "   messy   "
print(clean_me.strip())  # "messy" (Removes extra spaces)

# Splitting a sentence into a list of words
sentence = "Apple,Banana,Cherry"
print(sentence.split(",")) # ['Apple', 'Banana', 'Cherry']
```

## 2) Operators: The Math & Logic Tools

"Operators are the symbols that tell Python to perform a specific task. They are the 'verbs' of our code."

### A. Arithmetic & Assignment (The Accountant)

```python
# Basic Math
print(10 % 3)  # Modulus: Shows the remainder (1)
print(2 ** 3) # Exponent: 2 to the power of 3 (8)

# Assignment: Shorthand for updating a value
wallet = 100
wallet += 50  # Same as wallet = wallet + 50
print(wallet) # 150
```

### B. Comparison & Logical (The Security Guard)

These operators ask questions. The answer is always 'True' or 'False'.

```python
# Comparison
age = 18
print(age >= 18) # True

# Logical (Combining questions)
# "Are you 18 AND do you have a ticket?"
has_ticket = True
print(age >= 18 and has_ticket) # True
```

### C. Membership & Identity (The Detective)

Membership (in) checks if something is inside a collection. Identity (is) checks if two things are exactly the same object in memory.

```python
# Membership: Is 'a' in the word 'Apple'?
print("a" in "Apple") # False (Python is case-sensitive!)
print("A" in "Apple") # True

# Identity
x = [1, 2]
y = [1, 2]
print(x == y) # True (The contents are the same)
print(x is y) # False (They are two different "boxes" in memory)
```

# Module 4: Data Types & Collections

## 1) Lists: "The Organized Shelf" (Mutable & Ordered)

A List is like a shelf where you can keep anything—books, toys, or even other lists. The best part? You can change what's on the shelf anytime (Mutable) and the order never changes unless you move things (Ordered)

```python
# 1. Creating a list
students = ["Alice", "Bob", "Charlie"]
print("Original List:", students)


# 2. Accessing (Indexing & Slicing)
print("First student:", students[0])
print("Last two:", students[1:])

# 3. Adding items (.append)
students.append("David")
print("After adding David:", students)

# 4. Removing items (.pop)
removed_student = students.pop(1) # Removes Bob
print("Removed:", removed_student)
print("Final List:", students)

# 5. Lists can hold different types!
mixed_list = [10, "Hello", 3.14, True]
print("Mixed list:", mixed_list)
```

## 2) Tuples: "The Sealed Vault" (Immutable & Ordered)

A Tuple is like a list, but it's been 'Super-Glued.' Once you create it, you cannot change, add, or remove items (Immutable). We use these for things that should never change, like coordinates or a birthday.

```python
# 1. Creating a tuple (Uses parentheses)
location = (12.97, 77.59) # Latitude and Longitude of Bengaluru
print("Location:", location)

# 2. Accessing works just like lists
print("Latitude:", location[0])

# 3. PROOF of Immutability (This will cause an error!)
try:
    location[0] = 13.00
except TypeError as e:
    print("Error caught:", e) # You can't change a tuple!
```

## 3) Sets: "The Party Guest List" (Unordered & No Duplicates)

A Set is a collection where every item must be unique. If you try to add the same person to the party twice, the set just ignores the second one. Also, people in a set don't stand in a line; they move around (Unordered)

```python
# 1. Creating a set
ids = {101, 102, 103, 101, 101}
print("Set only keeps unique values:", ids) # {101, 102, 103}
#ids.pop()

# 2. Adding and Removing
ids.add(104)
ids.remove(102)
print("Updated set:", ids)


# 3. Sets are great for checking membership quickly
print(101 in ids) # True
```

## 4. Dictionaries: "The Phone Book" (Key-Value Pairs)

A Dictionary is the most powerful tool. Instead of numbers (0, 1, 2), we use labels (Keys) to find information (Values). Just like looking up a name in a phone book to find a number.

```python
# 1. Creating a dictionary {Key: Value}
user_profile = {
    "name": "Shreyash",
    "role": "Python Tutor",
    "experience": 10
}

# 2. Accessing values using keys
print("Name:", user_profile["name"])

# 3. Adding or Updating
user_profile["location"] = "Cloud"
user_profile["experience"] = 11 # Updating value
print("Updated Profile:", user_profile)

# 4. Removing a key
del user_profile["role"]
print("After deletion:", user_profile)

# 5. Getting all keys or values
print("Keys:", list(user_profile.keys()))
print("Values:", list(user_profile.values()))
```

## 5. Mutability vs Immutability Summary

If you can change it after birth, it's Mutable (List, Set, Dict). If it's frozen in time, it's Immutable (String, Tuple, Int, Float).

```python
# Demonstrating String Immutability
name = "Python"
# name[0] = "J"  # This would fail!

# To "change" it, we must create a NEW string
new_name = "J" + name[1:]
print("Original:", name)
print("New string:", new_name)
```

#Module 5: Conditional Statements

## 1. The **if** Statement:

Think of the if statement as a security guard at a club. He only lets you in if you meet a specific requirement (like being 18+). If you don't meet it, he says nothing and you just stand there.

Syntax:
   
   if condition:
     code to execute if condition is True.

```python
# Situation: Checking if a student passed
marks = 85

if marks >= 40:
    print("Congratulations! You passed.")

# Common Error: Forgetting the colon (:) or indentation
# if marks >= 40  <-- This will cause a SyntaxError
# print("Passed") <-- This will cause an IndentationError
```

## 2. The **if-else** Statement:

Now, imagine a GPS. If you take the correct turn, it says 'Continue straight.' Otherwise (else), it says 'Recalculating.' There is no middle ground; it's one or the other.

Syntax:

if condition:
    # runs if True

else:
    # runs if False

```python
# Situation: Even or Odd number checker
number = 7

if number % 2 == 0:
    print(f"{number} is Even.")
else:
    print(f"{number} is Odd.")
```

## 3. The if-elif-else Ladder: "The Multiple Choice"

What if there are many possibilities? Like a grading system (A, B, C, D, or F). We use elif (short for 'else if') to check multiple conditions one by one until we find a match."

  

Syntax:

if condition1:
    # code

elif condition2:
    # code

elif condition3:
    # code
    
else:
    # default code if nothing matches

```python
# Situation: Ticket pricing based on age
age = 25

if age < 5:
    print("Ticket is Free!")
elif age <= 18:
    print("Child ticket: $10")
elif age <= 60:
    print("Adult ticket: $20")
else:
    print("Senior ticket: $15")
```

## 4. Nested Conditions: "The Decision Tree"

Sometimes, one decision depends on another. It's like a locked door: First, you must have a key. Then, you must turn the handle. If you don't have the key, the handle doesn't even matter."

  

Syntax:

  if condition1:

    if condition2:
       # code runs only if both are True

```python
# Situation: ATM Withdrawal Logic
balance = 1000
withdrawal_amount = 500
pin_correct = True

if pin_correct:
    print("PIN accepted.")
    if withdrawal_amount <= balance:
        print("Withdrawal successful!")
    else:
        print("Insufficient funds.")
else:
    print("Incorrect PIN. Access denied.")
```

## 5. Short-hand (Ternary Operator): "The Quick Text"

This is for when you want to write a simple if-else in just one line. It’s like a quick 'Yes/No' text message instead of a formal letter."

  

Syntax:

value_if_true if condition else value_if_false

```python
# Situation: Assigning a status
rainy = True
action = "Take umbrella" if rainy else "Wear sunglasses"

print(action)
```

Common Pitfalls to Explain:
The Assignment vs. Comparison Trap:

if score = 100: (Error! This is trying to "set" the score).

if score == 100: (Correct! This is "checking" the score).

Order Matters: In an elif ladder, Python stops at the first True condition. If you check age < 60 before age < 18, an 8-year-old will be caught in the age < 60 block

# Module 6: Loops In Python


## 1. The for Loop: "The Inspection Line"

Imagine a conveyor belt in a chocolate factory. A for loop looks at every single item on that belt, one by one, until it reaches the end. We use this when we know exactly how many items we need to check (like a list of names or a range of numbers).

  

Syntax:

for item in sequence:
    # do something with item

```python
# Situation: Printing a countdown
# range(start, stop_before)
for i in range(1, 6):
    print(f"Counting: {i}")

# Situation: Iterating over a list
fruits = ["Apple", "Banana", "Cherry"]
for fruit in fruits:
    print(f"I want to eat a {fruit}")
```

## 2. The while Loop: "The Security Guard"

A while loop is different. It’s like a security guard standing at a gate saying, 'As long as your ticket is valid, you can keep dancing.' It repeats as long as a condition is True. But be careful! if the condition never becomes False, the guard stays there forever—this is an Infinite Loop!

  

Syntax:

    while condition:

      # do something

      # update the condition

```python
# Situation: A battery charging
battery = 80
while battery < 100:
    print(f"Battery at {battery}%, charging...")
    battery += 5  # Crucial: If we forget this, the loop never ends!

print("Fully Charged!")

# --- Common Error: The Infinite Loop ---
# count = 1
# while count > 0:
#    print("I will never stop!")
#    # (In Colab, you'd have to hit the 'Stop' button to kill this)
```

## 3. Iterating Over Different Collections

Python is smart. You can use a for loop to 'walk' through almost any collection we learned in Module 4.

```python
# A. Walking through a String
word = "AI"
for letter in word:
    print(f"Letter: {letter}")

# B. Walking through a Dictionary (The Phone Book)
# By default, it loops through KEYS.
user_scores = {"Alice": 90, "Bob": 85}

print("\n--- Dictionary Loop ---")
for name, score in user_scores.items():
    print(f"{name} scored {score} points")
```

## 4. Nested Loops: "The Clock Analogy"

A nested loop is a loop inside a loop. Think of a clock: The 'Seconds' hand must go around 60 times before the 'Minutes' hand moves just once. The inner loop completes its entire cycle for every single step of the outer loop.

```python
# Situation: Creating a coordinate grid (3x3)
for x in range(1, 4):      # Outer Loop
    for y in range(1, 4):  # Inner Loop
        print(f"Point: ({x}, {y})")
```

Common Pitfalls:

Off-by-One Error: Students often forget that range(1, 5) stops at 4.

Always remember: "Python stops at the gate, it doesn't go inside."

Forgetting to Update: In while loops, forgetting to increment your counter (i += 1) will crash your browser or computer.

Indentation: If the code isn't pushed to the right, it isn't "inside" the loop.

# Module 7: Control Statements

## 1. The break Statement: "The Emergency Exit"

Imagine you are searching for a specific book on a shelf of 1,000 books. Once you find it, do you keep looking at the rest? No! You stop. break tells Python to exit the loop immediately, even if the loop wasn't finished.

  

Syntax:

    for item in sequence:

      if condition:

          break

```python
# Situation: Finding a specific "Golden Ticket" in a box of chocolates
chocolates = ["Milk", "Dark", "Golden Ticket", "White", "Caramel"]

for chocolate in chocolates:
    if chocolate == "Golden Ticket":
        print("Found it! Stopping the search.")
        break  # The loop stops here; it won't check White or Caramel.
    print(f"Checking {chocolate}...")

# Example 2: Stopping an infinite loop
count = 0
while True: # This would normally run forever
    print(count)
    count += 1
    if count >= 5:
        break # Emergency brake!
```

## 2. The continue Statement: "The Skip Button"

Think of a playlist. If a song comes on that you don't like, you hit 'Next.' You don't throw the phone away; you just skip that one song and move to the next one. continue tells Python to stop the current turn and jump straight to the start of the next turn.

  

Syntax:

    for item in sequence:

      if condition:

         continue


```python
# Situation: Printing only odd numbers (skipping evens)
for num in range(1, 10):
    if num % 2 == 0:
        continue # Skip the rest of this code block for even numbers
    print(f"Odd number: {num}")

# Situation: Skipping "Hidden" files in a list
files = ["photo.jpg", ".DS_Store", "resume.pdf", ".config"]
for f in files:
    if f.startswith("."):
        continue # Don't process hidden system files
    print(f"Processing: {f}")
```

## 3. The pass Statement: "The Placeholder"

Sometimes you know you need a loop or a function, but you haven't written the logic yet. Python hates empty blocks of code—it will crash! pass is a way of saying 'I'm working on this, please don't complain yet.' It does absolutely nothing.

  

Syntax:

    if condition:

    pass # To be implemented later

```python
# Situation: Planning a complex game logic
is_game_over = False

if is_game_over:
    # I'll write the "Game Over" screen logic tomorrow
    pass
else:
    print("Game is still running!")

# Using pass in a loop
for i in range(5):
    if i == 3:
        pass # I might want to do something special for '3' later
    print(i)
```

### Summary:

The "Traffic Light" Analogy

break: The Red Light. Stop everything and go home.

continue: The Yellow Light. Skip this intersection and go to the next one.

pass: The Green Light on an empty road. Just keep going, nothing to see here.

# Module 8: Functions

## 1. What is a Function? "The Juicer Analogy"

Imagine you have a fruit juicer. You put fruit in (Arguments), the machine does the work (Logic), and it pours out juice (Return Value). You don't care how the gears inside turn; you just care that every time you give it an orange, you get orange juice. That’s a function—a reusable machine for your code!"

  

Syntax:

    def function_name(parameters):

    # logic goes here

    return result

## 2. Defining & Calling Functions

Defining a function is like writing a recipe. It doesn't cook the food yet; it just stays in the book. Calling the function is when you actually start cooking.

```python
# 1. Defining the "Recipe"
def greet_student(name):
    print(f"Hello {name}, welcome to Python class!")

# 2. Calling the "Recipe" (Executing it)
greet_student("Alice")
greet_student("Bob")
```

## 3. Parameters vs. Arguments

These terms are often confused. Think of Parameters as the 'Placeholders' (the empty slots in the juicer) and Arguments as the 'Actual Values' (the fruit you drop in).

```python
def add_numbers(a, b): # a and b are Parameters
    print(a + b)

add_numbers(10, 5) # 10 and 5 are Arguments
```

## 4. The return Statement: "The Delivery Driver"

A function can either just 'do' something (like print) or it can 'give back' something. return is like a delivery driver who brings the result back to you so you can store it in a variable.

```python
def calculate_area(length, width):
    return length * width

# We store the 'returned' value in a variable
my_room_area = calculate_area(10, 12)
print(f"My room is {my_room_area} sqft.")
```

## 5. Default & Keyword Arguments

Sometimes, we have a 'default' setting. Like a coffee machine that defaults to 'Medium' sugar unless you tell it otherwise.

```python
def make_coffee(sugar="Medium"):
    print(f"Making coffee with {sugar} sugar.")

make_coffee()          # Uses default: Medium
make_coffee("Extra")   # Overrides default: Extra

# Keyword Arguments: Specifying which value belongs to which parameter
def profile(name, age):
    print(f"Name: {name}, Age: {age}")

profile(age=25, name="John") # Order doesn't matter if you use names!
```

## 6. Special Tools: *args and kwargs

What if you don't know how many items a student will bring to class? *args lets you accept a long list of items, and kwargs lets you accept a whole dictionary of labeled items.

```python
# *args: Accepts any number of values as a TUPLE
def sum_all(*numbers):
    return sum(numbers)

print(sum_all(1, 2, 3, 4, 5)) # Output: 15

# **kwargs: Accepts any number of labeled values as a DICTIONARY
def show_info(**data):
    for key, value in data.items():
        print(f"{key}: {value}")

show_info(City="Bengaluru", Weather="Sunny", Temp=30)
```

```python
def make_pizza(size, *toppings):
    print(f"Making a {size} inch pizza with the following toppings:")
    for topping in toppings:
        print(f"- {topping}")

make_pizza(12,"Pepperoni", "Mushrooms", "Extra Cheese")
```

Common Errors:

Scope Error: Trying to use a variable inside a function from the outside. "What happens in the function, stays in the function!"

Missing Return: Forgetting return and getting None when trying to save a result.

Indentation: Remember that the entire "body" of the function must be indented.

## 7) List Comprehension

Imagine you have a basket of 10 apples, and you want to put only the 'Red' ones into a new basket. Normally, you’d need a for-loop, an if-statement, and an .append() command. List Comprehension lets you do all of that in just one line of code. It’s the 'Shortcut Lane' of Python!

Syntax:

    new_list = [expression for item in iterable if condition]

Expression: What do we want to save? (The 'Red Apple')

For Item in Iterable: Where are we looking? (The 'Big Basket')

If Condition: (Optional) Is there a filter? (Only 'Red' ones)

**Situation A:**

```python
#Let's take a list of numbers and create a new list that squares each number.
numbers = [1, 2, 3, 4, 5]

# The "Old" way (4 lines)
squares_old = []
for n in numbers:
    squares_old.append(n**2)
print("Old Way:", squares_old)

# The "Pythonic" way (1 line)
squares_new = [n**2 for n in numbers]
print("List Comprehension:", squares_new)
```

**Situation B:**

 Adding a Filter (The 'if' condition)
"Now, what if we only want to square the EVEN numbers and ignore the odd ones?"

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

# [Result | Loop | Filter]
even_squares = [n**2 for n in numbers if n % 2 == 0]

print("Only Even Squares:", even_squares)
```

**Situation C:**

Working with Strings
"We can even use it to clean up data, like making names uppercase."

```python
students = ["alice", "bob", "charlie"]

# Transformation: "Make s uppercase for every s in students"
upper_students = [s.upper() for s in students]

print(upper_students) # ['ALICE', 'BOB', 'CHARLIE']
```

**Common Errors & Situations:**

1) Complexity Overload: If the logic is too long (like 3 or 4 nested steps), don't use list comprehension. It becomes unreadable. "If it doesn't fit on one line comfortably, use a regular loop."

2) The Result is ALWAYS a List: Even if you only find one item, it will be inside [].

# Module 9: Lambda Functions

## 1. What is a Lambda? "The Sticky Note"

If a regular function is a 'Handwritten Recipe Book,' a Lambda function is a 'Sticky Note.' It’s small, contains only one instruction, and you usually throw it away after one use. We call these Anonymous Functions because they don't even need a name!"

  

Syntax:

    lambda arguments : expression

Keyword: lambda

Arguments: The inputs (like $x, y$).

Expression: The single piece of math or logic to return.


```python
#Let's compare a regular function that doubles a number with a Lambda version. Look how much space we save!
# Regular Function
def double(x):
    return x * 2

# Lambda Function (The Shortcut)
double_lambda = lambda x: x * 2

print(double(5))        # 10
print(double_lambda(5)) # 10
```

```python
increment_by_i = [lambda x: x + i for i in range(10)]
print( increment_by_i [3](3))
```

## 2) Advanced Use Cases of Lambda (The Power Tools)

Lambdas are rarely used by themselves. Their real power comes when they work inside other functions like map() and filter().

  

A. The map() Function: "The Assembly Line"

Use map() when we want to apply a rule to every single item in a list.

```python
prices = [10, 20, 30, 40]
# Use case: Adding 18% tax to every price
taxed_prices = list(map(lambda p: p * 1.18, prices))

print("Prices with Tax:", taxed_prices)
```

B. The filter() Function: "The Security Guard"

Use filter() when you want to keep only the items that meet a certain condition.

```python
ages = [12, 18, 25, 8, 30, 15]
# Use case: Keeping only "Adult" ages (18+)
adults = list(filter(lambda a: a >= 18, ages))

print("Adults only:", adults)
```

Common Errors:

Multiple Expressions: We often try to put two things inside a lambda (e.g., lambda x: x+1; print(x)). Error! A lambda can only do one single thing.

Forgetting list(): When using map or filter, Python returns a "Map Object" (a memory location). You must wrap it in list() to actually see the results!

C. The sorted() Custom Sort: "The Librarian"

This is a professional-level use case! Imagine we have a list of names and we want to sort them by the length of the name, not alphabetically.

```python
names = ["Alice", "Bob", "Charlie", "David"]

# Sort by the length of each name
sorted_names = sorted(names,key = lambda name: len(name))

print("Sorted by length:", sorted_names)
```

now what if we have a list of students with their marks, and we want to find out who came first? We use a lambda to tell Python: 'Don't look at the name, look at the score!'

```python
# A list of dictionaries
students = [
    {"name": "Alice", "score": 88},
    {"name": "Bob", "score": 95},
    {"name": "Charlie", "score": 78}
]

# Sort by 'score' in descending order (highest first)
top_students = sorted(students, key=lambda s: s['score'], reverse=True)

print("Rank List:")
for s in top_students:
    print(f"{s['name']}: {s['score']}")
```

D.The reduce() Function: "The Rolling Snowball"

Imagine we have a list of numbers and we want to squish them down into a single final value—like multiplying all of them together. reduce() takes the first two items, applies the lambda, takes that result, applies it to the third item, and so on. It’s like a rolling snowball getting bigger and bigger!"

Note: we must import this from functools.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Use case: Finding the product of all numbers (1*2*3*4*5)
product = reduce(lambda x, y: x * y, numbers)

print("The total product is:", product) # Output: 120
```

E. The min() and max() with Key: "The Extremes"

Similar to sorting, sometimes we just want the single highest or lowest value in a complex list. Instead of writing a loop to compare everything, we use a Lambda as a 'search criteria'.

```python
# Situation: Finding the cheapest fruit in a store
inventory = [("Apple", 50), ("Banana", 20), ("Cherry", 100)]

# We tell Python to look at index 1 (the price)
cheapest = min(inventory, key=lambda item: item[1])
most_expensive = max(inventory, key=lambda item: item[1])

print(f"The cheapest fruit is {cheapest[0]} at {cheapest[1]} coins.")
print(f"The most expensive is {most_expensive[0]} at {most_expensive[1]} coins.")
```

# Module 10: Error Handling

## 1. The try-except Block: "The Safety Net"
Imagine you are a trapeze artist. You hope you’ll catch the bar, but just in case you don’t, you have a Safety Net below. In Python, try is the performance, and except is the safety net that catches the program before it hits the ground (crashes)."

  

Syntax:

    try:
    # Code that might cause an error
    
    except ErrorType:
    # What to do if the error happens

```python
# Situation 1: Dividing by Zero
try:
    number = int(input("Enter a number to divide 100 by: "))
    result = 100 / number
    print(f"Result is {result}")
except ZeroDivisionError:
    print("Oops! You can't divide by zero. Try a different number.")
except ValueError:
    print("That wasn't a number! Please enter digits only.")
```

## 2. The finally Block: "The Cleanup Crew"

Think of a kitchen. Whether you successfully bake a cake or you accidentally burn it, you must wash the dishes. finally is the part of your code that runs no matter what happened in the 'try' block. It’s perfect for closing files or turning off database connections."

  

Syntax:

Python

    try:
    # Action
    except:
    # Error handling
    finally:
    # Always runs (Cleanup)

```python
try:
    print("Opening the oven...")
    # Imagine something goes wrong here
    result = 10 / 0
except:
    print("Something went wrong in the oven!")
finally:
    print("Closing the oven door and cleaning up.") # This always prints!
```

## 3. Custom Exceptions: "The Rules of the House"

Sometimes, Python doesn't think something is an error, but you do. For example, if a user enters an age of -5, Python is fine with the math, but your 'House Rules' say age must be positive. We use raise to trigger our own custom warnings.

  

Syntax:

    if condition_is_bad:
    raise Exception("Your custom message")

```python
# Situation: Validating age for a voting app
age = -10

try:
    if age < 0:
        raise ValueError("Age cannot be negative!")
    elif age < 18:
        print("Too young to vote.")
    else:
        print("You can vote!")
except ValueError as e:
    print(f"Input Error: {e}")
```

## 5: Encapsulation

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    # Getter
    def get_balance(self):
        return self.__balance

    # Setter
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

acc = BankAccount(1000)
acc.deposit(500)
print(f"Current Balance: {acc.get_balance()}")
print(f"Current Balance: {acc._BankAccount__balance}") #Private variable can be accessed using this "Mangled Name" Method althought its not a good and ethical Practice to do so
#print(f"Current Balance: {acc._balance}") # "Mangled Name"method to access the Protected Variable.


```

## 6: Inheritance

### Parent and Child Classes
Just like you might inherit your parents' eye color, a **Child Class** inherits all the tools and data from a **Parent Class**. This prevents us from writing the same code twice!

**Types:** Single, Multiple, Multilevel, Hierarchical.

```python
class Animal: # Parent
    def breathe(self):
        print("Breathing...")

class Dog(Animal): # Child (Single Inheritance)
    def bark(self):
        print("Woof! Woof!")

my_dog = Dog()
my_dog.breathe() # Inherited
my_dog.bark()
```

```python
#Single InheritanceS
class Parent:
    def feature1(self):
        print("Feature 1 from Parent")

class Child(Parent):
    def feature2(self):
        print("Feature 2 from Child")

obj = Child()
obj.feature1()  # Search: Child -> Parent (Found)
obj.feature2()  # Search: Child (Found)
```

```python
# Multiple Inheritance
class Father:
    def skills(self):
        print("Father's Gardening skills")

class Mother:
    def skills(self):
        print("Mother's Cooking skills")

class Child(Father, Mother):  # Father is first priority
    pass

c = Child()
c.skills()  # Output: Father's Gardening skills (due to MRO)
print(Child.__mro__)
```

```python
#Multi-Level Inheritance
class Grandfather:
    def land(self):
        print("Grandfather's 50 acres")

class Father(Grandfather):
    def house(self):
        print("Father's Mansion")

class Son(Father):
    def car(self):
        print("Son's Sports Car")

s = Son()
s.land()   # Inherited from Grandfather through Father
s.house()  # Inherited from Father
```

```python
# Hierarchical Inheritance
class Employee:
    def company_policy(self):
        print("Company Policy: 9 to 5")

class Developer(Employee):
    def code(self):
        print("Coding...")

class Designer(Employee):
    def draw(self):
        print("Designing...")

dev = Developer()
dsgn = Designer()

dev.company_policy() # Common parent
dsgn.company_policy() # Common parent
dev.draw() #This would raise an AttributeError
```

## 7: Polymorphism

### One Name, Many Forms
The word 'Cut' means something different to a director, a barber, and a tailor. In code, **Polymorphism** allows the same function name to behave differently depending on who is using it.

* **Overriding:** Child changes a Parent's method.
* **Duck Typing:** 'If it walks like a duck and quacks like a duck, it's a duck!'

In the simplest terms, Polymorphism is a Greek word that means "Many Forms" (Poly = Many, Morph = Forms). In Python, it is the ability of different objects to be treated as instances of the same general class through the same interface, even though they each behave differently.

The Logic Behind the Execution
When you call a method on an object, Python doesn't look at the type of variable you are using; it looks at the type of the object itself. This is the core of dynamic execution in Python.



```python
class Bird:
    def fly(self):
        print("Most birds can fly.")

class Penguin(Bird):
    def fly(self): # Overriding
        print("Penguins cannot fly, but they can swim!")

def check_flight(bird):
    bird.fly()

p = Penguin()
check_flight(p)
```

1. Method Overriding (Run-time Polymorphism)
This happens in Inheritance. A child class can provide a specific implementation of a method that is already provided by its parent class.

The Flow: When you call the method, Python checks the child class first. If it finds the method there, it executes it and stops—ignoring the parent's version.

```python
#Method Overriding
class Bird:
    def fly(self):
        print("Most birds can fly.")

class Penguin(Bird):
    # Overriding the parent method
    def fly(self):
        print("Penguins cannot fly, they swim!")

class Eagle(Bird):
    pass # Uses the parent's version

# Execution Flow
birds = [Eagle(), Penguin()]

for b in birds:
    b.fly()
    # For Eagle: Search Eagle -> Found nothing -> Search Bird -> Found (Prints 'Most birds fly')
    # For Penguin: Search Penguin -> Found (Prints 'Penguins swim')
```

2. Duck Typing (The "Pythonic" Polymorphism)
Python follows the philosophy: "If it walks like a duck and quacks like a duck, then it is a duck."

The Logic: Python doesn't care about the "Class" of an object. It only cares if the object has the Method you are trying to call. If the method exists, Python runs it.

```python
#Duck Typing
class Laptop:
    def code(self):
        print("Coding on a laptop...")

class Tablet:
    def code(self):
        print("Coding on a tablet with a keyboard...")

class Developer:
    def work(self, device):
        device.code() # Python doesn't care what 'device' is, as long as it has .code()
        print("Coding on a PC with a keyboard...")


dev = Developer()
dev.work(Laptop()) # Works
dev.work(Tablet()) # Works
#dev.work
```

3. Method Overloading (Compile-time Polymorphism)
In languages like Java or C++, you can have multiple methods with the same name but different parameters. Python does not support this natively.

The Logic: In Python, the last defined method with a specific name replaces any previous ones. However, we simulate overloading using Default Arguments or *args.

```python
#Method Overloading
class Math:
    # Simulating overloading
    def add(self, a, b, c=0):
        return a + b + c

m = Math()
print(m.add(5, 10))     # Works with 2 arguments
print(m.add(5, 10, 20)) # Works with 3 arguments
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        # 'self' is the first point, 'other' is the second point
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Point(new_x, new_y)

    # Overloading the 'print' behavior so we can see the result
    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)

# The logic: This line internally calls p1.__add__(p2)
p3 = p1 + p2

print(f"Point 1: {p1}")
print(f"Point 2: {p2}")
print(f"Result of p1 + p2: {p3}")
```

## 8: Abstraction

### Hiding Complexity
When you drive a car, you press the brake pedal. You don't need to know how the hydraulic fluid moves. **Abstraction** means showing only the essential features and hiding the complicated 'how-it-works' logic."

* **Abstract Class:** A blueprint that *forces* children to implement certain methods.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side * self.side

sq = Square(5)
print(f"Square Area: {sq.area()}")
```

### -> Detailed Explaination:

Abstraction is the process of hiding the complex internal implementation details of a system and showing only the necessary features to the user.

The Logic: "The Universal Interface"
Think of a TV Remote. You know that pressing the 'Power' button turns the TV on. You don't need to know about the infrared signals, the circuit boards, or the electricity flowing inside. The button is an Abstract Interface.

In code, Abstraction allows us to define what an object should do without specifying how it does it. This is handled in Python using Abstract Base Classes (ABC).

1. Abstract Classes and Methods
An Abstract Class is a class that cannot be instantiated (you cannot create an object from it). It exists only to serve as a blueprint for other classes.

An Abstract Method is a method that has a declaration but no implementation. Any child class inheriting from this abstract class must provide the logic for that method, or Python will throw an error.

2. The Flow: How Python Enforces Abstraction
To use abstraction, we must import the abc module. When Python sees the @abstractmethod decorator, it "locks" the class. It won't let you create an object until all abstract methods are defined in a child class.

```python
from abc import ABC, abstractmethod

# The Blueprint (Abstract Class)
class PaymentGateWay(ABC):

    @abstractmethod
    def process_payment(self, amount):
        # No logic here, just a requirement
        pass

    def generate_receipt(self):
        # Abstract classes can still have normal methods!
        print("Generating universal receipt...")

# The Implementation (Child Classes)
class PayPal(PaymentGateWay):
    def process_payment(self, amount):
        print(f"Processing ${amount} via PayPal Secure Servers.")

class Stripe(PaymentGateWay):
    def process_payment(self, amount):
        print(f"Processing ${amount} via Stripe API.")

# --- Execution Logic ---
#p = PaymentGateWay() # This would throw a TypeError!

payment1 = PayPal()
payment1.process_payment(100) # Runs PayPal logic
payment1.generate_receipt()   # Runs shared logic
```
