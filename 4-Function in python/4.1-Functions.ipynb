```python
# Syntax Of function 
def fun_name(parameters):
    """Docstring""" #doc string is just to used to, write some kind of comments about what the specific function does 
                    # so that any developer will be able to understand it.
    #function Body
    return expression
```

```python
# Why we require functions ?
num =24
if num%2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
# If we have top reuse the same piece of code 
# we can do it with the help of functions without writing the Function all over again
# Defining the function->
def even_odd(num):
    """This Functions Checks whether a number is even or odd"""
    if num%2 == 0:
        print("The number is Even")
    else:
        print("The number is Odd")
```

```python
#Calling the Function
even_odd(24)
```

```python
## Default Parameters
def greet(name):
    print(f"Hello {name}, Welcome to Cybage")

greet("Shrey")
```

```python
def greet(name):
    print(f"Hello {name}, Welcome to Cybage")

greet()
```

```python
def greet(name= "Employee"): # setting the default parameter 
    print(f"Hello {name}, Welcome to Cybage")

greet()
```

```python
def greet(name= "Employee"): # setting the default parameter 
    print(f"Hello {name}, Welcome to Cybage")

greet("Shrey")
```

```python
## Variable Length Arguments 
#Positional and Keywords Arguments 
#Postional arguments are denoted by '*'
def print_num(*Shrey):
    for number in Shrey:
        print(number)

print_num(1,2,3,4,5,"Shrey")
```

```python
def print_num(*args): #Tradinal way of notation is *args
    for number in args:
        print(number)

print_num(1,2,3,4,5,"Shrey")
```

```python
#Keyword Arguments are denoted by '**'
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name ="Shrey", age ="23",country = "India")
```

```python
# Passing Both Postional and Keyword Arguments in the same Function
def print_details(*args,**kwargs):
    for val in args:
        print(f" Postional argument :{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,"Shrey",name ="Shrey", age ="23",country = "India")
```

```python
## Return statements 
def multiply(a,b):
    return a*b

multiply(2,3)
```
