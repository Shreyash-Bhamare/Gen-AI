## Introduction to Tuples

Explaination:

Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
## Creating a tuple 

empty_tuple = ()
print(empty_tuple) 
print(type(empty_tuple))
```

```python
lst =list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
num = tuple([1,2,3,4,5])
num
```

```python
list((1,2,3,4,5))
```

```python
mixed_tuple = (1, "Hello",3.14, True)
print(mixed_tuple)
```

```python
# Accessing Tuple Elements
num
```

```python
print(num[0])
print(num[2])
print(num[-1])


```

```python
print(num[1:])
print(num[1:4])
print(num[:5])
print(num[::])
print(num[::2])
print(num[::-1])

```

```python
## Packing and Unpacking of Tuples
#Packing
pack_tuple = 1, "Hello", 3.14, False
print(pack_tuple)
print(type(pack_tuple))
```

```python
#unpacking
a,b,c,d = pack_tuple

print(a)
print(b)
print(c)
print(d)
```

```python
##Unpacking Using *
print(num)
first, *middle,last=num
print(first)
print(middle)
print(last)
```

```python
## Nested List 
lst = [[1,2,3,4,5],[6,7,8,9],[1,"Hello",3.14,"c"]]
print(lst)
```

```python
## Acccesing a nested list element
print(lst[0],lst[0][2])
```

```python
## Slicing In Nested List 
print(lst[2][0:3])
```

```python
## Nested Tuple
nested_tuple = (1, 2, (3,4), (5,6,(7,8)))
print(nested_tuple)

## Acessing Nested Tuple Elements
print(nested_tuple[2])
```

```python
print(nested_tuple[2][1])
```

```python
nested_tuple = ( (3,4), (5,6,(7,8)))
print(nested_tuple)
```

```python
## iterating over Tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```
