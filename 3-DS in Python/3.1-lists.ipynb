#### List Comprehension

1) Basic Syntax ->                [expression for item in iterable]
2) with conditional logic  ->     [expression for item in iterable if condition]
3) Nested List Comprehension ->   [expression for item1 in iterable1 for item2 in iterable2]


```python
## Basic List Comphrehension

square = [num**2 for num in range (10)] 
```

```python
print(square)
```

```python
# List Compherension with condition 
lst=[]
for i in range (10):
    if i%2 == 0:
        lst.append(i)

print(lst)        
```

```python
even_n = [num for num in range (10) if num%2==0]
```

```python
print(even_n)
print(type(even_n))
```

```python
#Nested List Comprehension
num1 = [1,2,3,4]
num2 = ['a', 'b','c', 'd']

pair  = [[i,j] for i in num1 for j in num2]

print(pair )
```

```python
##List Comprehensionwith function calls 
words = ['hello', 'world', 'python', 'is', 'awesome']
lengths = [len(word) for word in words]
print(lengths)
```

### Conclusion 

List comprehension are powerful and concise way to create a list in python. They are syntactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehension will help you write cleaner and more efficient pyhton code. 

# Real-World Examples using Lists in pyhton 


### Example 1: Manage a to-do-list of tasks 


<b> $ Create a To-Do-List to keet track of task <b>

```python
to_do_list = ["Buy Groceries","Clean the house","Pay Bills"]

##Adding more tasks
to_do_list.append("Schedule a meeting")
to_do_list.append("Go to GYM")

##Removing a Complete task
to_do_list.remove("Clean the house")

## Checking if task is in the list
if "Pay Bills" in to_do_list:
     print("Don't forget to pay bills today!")

##Printing all tasks
print("Remaining Tasks on Your To-Do List: ")
for task in to_do_list:
    print(f" -{task}")
```

# Example 2: Organizing Student Grades 

<b>$ Create a list to store and calculate the avg grades of students.<b>

```python
# Organizing Student Grades
student_grades = [85, 92, 78, 90, 88]

# Adding a new grade
student_grades.append(95)

# Calculating the average grade
average_grade = sum(student_grades) / len(student_grades)
print(f"Average Grade: {average_grade}")

#Finding the highest and lowest grades
highest_grade = max(student_grades) 
lowest_grade = min(student_grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

# Example 3: Managing An Inventory 

<b>$ Use a list to manage inventory items in a store <b>

```python
# Managing An Inventory
inventory = ["Ducati", "BMW", "KTM", "Honda", "Yamaha"]
print("Inventory List: ")
for item in inventory:
    print(f" - {item}")

# Adding new items to the inventory
inventory.append("Suzuki")
inventory.append("Harley-Davidson")

# Removing an item from the inventory
inventory.remove("Honda")

#Checking if an item is in stock
item = "KTM"
if item in inventory:
    print(f"{item} is in stock.")
else:
    print(f"{item} is out stock.")

#Printing the Inventory after updates 
print("Inventory List after Ops: ")
for item in inventory:
    print(f" - {item}")
```

# Example 4: Collecting the User 

<b>$ Use a list to collect and analyse user feedback <b>

```python
# Collecting User Feedback
user_feedback = ["Great service!", "Could improve delivery time.", "Excellent product quality.", "Average experience overall.", "Will shop again!"]

#Adding new feedback
user_feedback.append("Customer support was very helpful.")

# Counting specific feedback
positive_feedback_count = sum(1 for feedback in user_feedback if "Great" in feedback.lower() or "Excellent" in feedback.lower() or "again" in feedback.lower() or "helpful" in feedback.lower())
print(f"Number of Positive Feedbacks: {positive_feedback_count}")

#Printing all feedback
print("User Feedbacks: ")
for feedback in user_feedback:
    print(f" - {feedback}")
```
